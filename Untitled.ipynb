# Pipeline

I will use the feature gender to identify if gender has or has not an influence on academic performance. 
However, as our feature is a string(text) we have to apply one hot encoder to continue our process and we will also apply StandardScaler only in our numeric features. 

I created a pipeline to apply OneHotEncoder in my categorical feature and StandardScaler in my numeric features as we can see in the 72 cells the dataset is composed by gender_Female which is 0.0 or 1.0 means that 0 = female and 1 = male. Our numeric features were StandardScaler [Click here to visualize the Dataframe](#hear_new_df) 
    


```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
```

```python
cate_feat = ['gender']
numeric_feat = ['math_score', 'reading_score', 'writing_score' ]
```

```python
preprocessor = ColumnTransformer([
            ('onehot', OneHotEncoder(drop='first'), cate_feat), 
            ('scaler', StandardScaler(), numeric_feat )

    
])
```

```python
transf_data = preprocessor.fit_transform(df1)
```

```python
df2_pipe = pd.DataFrame(transf_data, columns=['gender_Female', 'math_score', 'reading_score', 'writing_score'])
```
