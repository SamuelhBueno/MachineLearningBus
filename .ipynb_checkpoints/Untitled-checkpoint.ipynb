# Pipeline

I will use the feature gender to identify if gender has or has not an influence on academic performance. 
However, as our feature is a string(text) we have to apply one hot encoder to continue our process and we will also apply StandardScaler only in our numeric features. 

I created a pipeline to apply OneHotEncoder in my categorical feature and StandardScaler in my numeric features as we can see in the 72 cells the dataset is composed by gender_Female which is 0.0 or 1.0 means that 0 = female and 1 = male. Our numeric features were StandardScaler [Click here to visualize the Dataframe](#hear_new_df) 
    


```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
```

```python
cate_feat = ['gender']
numeric_feat = ['math_score', 'reading_score', 'writing_score' ]
```

```python
preprocessor = ColumnTransformer([
            ('onehot', OneHotEncoder(drop='first'), cate_feat), 
            ('scaler', StandardScaler(), numeric_feat )

    
])
```

```python
transf_data = preprocessor.fit_transform(df1)
```

```python
df2_pipe = pd.DataFrame(transf_data, columns=['gender_Female', 'math_score', 'reading_score', 'writing_score'])
```

# We are set up to start to apply cluster in our dataset, before we continue let's plot a scatter plot

# K-MEANS

```python
from sklearn.cluster import KMeans
```

```python
wcss = []  
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, max_iter = 300, n_init = 10, random_state = 0)   
                                                                                    
    kmeans.fit(df2_pipe)                                                                   
    wcss.append(kmeans.inertia_)                                                   
```

```python
plt.plot(range(1, 11), wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')    # within cluster sum of squares
plt.show()
```

# Explaining the plot: 

- The plot shows the elbow points are around 3 or 4.




```python
kmeans = KMeans(n_clusters=3, random_state=42)
```

```python
clusters = kmeans.fit_predict(df2_pipe)
```

```python
df2_pipe['cluster'] = clusters
```

```python
fig = px.scatter(df2_pipe, x='reading_score', y='writing_score', color='cluster',
                 title='Scatter Plot', labels={'reading_score': 'Reading Score',
                                                'writing_score': 'Writing Score'}, 
                 width=1000, height=800)
# Show the plot
fig.show()
```

```python
fig = px.scatter(df2_pipe, x='math_score', y='reading_score', color='cluster',
                 title='Scatter Plot', labels={'reading_score': 'Reading Score',
                                                'writing_score': 'Writing Score'}, 
                 width=1000, height=800)
# Show the plot
fig.show()
```

# Hierarchical clustering

```python
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
compl_clust= linkage(df2_pipe, method="complete", metric="euclidean")
avg_clust = linkage(df2_pipe, method="average", metric="euclidean")
singl_clus = linkage(df2_pipe, method="single", metric="euclidean")
```

```python
dendrogram(compl_clust)
plt.show()
```

```python
dendrogram(avg_clust)
plt.show()
```

```python
dendrogram(singl_clus)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering


hc = AgglomerativeClustering(n_clusters = 3, affinity = 'euclidean', linkage = 'complete')

# Train the model by calling a method 'fit_predict(X)'
y_hc = hc.fit_predict(df2_pipe)
```

```python
plt.scatter(df2_pipe.iloc[y_hc == 0, 0], df2_pipe.iloc[y_hc == 0, 1], s=50, c='red', label='Cluster 1')
plt.scatter(df2_pipe.iloc[y_hc == 1, 0], df2_pipe.iloc[y_hc == 1, 1], s=50, c='blue', label='Cluster 2')
plt.scatter(df2_pipe.iloc[y_hc == 2, 0], df2_pipe.iloc[y_hc == 2, 1], s=50, c='green', label='Cluster 3')
plt.title('Clusters of customers')
plt.xlabel('X-axis Label')  # Replace with appropriate label
plt.ylabel('Y-axis Label')  # Replace with appropriate label
plt.legend()
plt.show()
```
